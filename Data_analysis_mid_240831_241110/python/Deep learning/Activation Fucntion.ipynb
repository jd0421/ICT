# 활성화 함수
- 각 함수는 입력 값을 특정 방식으로 변환하여 비선형성을 추가

### 활성화 함수의 종류
  - ReLU(ReLU 계열 : ELU GelU swift  기타등등)
  - 시그모이드
  - 하이퍼볼릭탄젠트
  - 소프트맥스

```python
import numpy as np
import matplotlib.pyplot as plt

# 입력 값 범위
x = np.linspace(-10, 10, 400)
```

# ReLU
  - 수식: 
$$
\operatorname{ReLU}(x)=\max (0, x)
$$


  - 출력 범위: [0, ∞)
  - 설명:   
  > 입력 값이 양수일 경우 그대로 출력하고, 음수일 경우 0을 출력합니다.   
  계산이 간단하고 그래디언트 소실 문제가 적어 최근 많이 사용되는 활성화 함수입니다.

```python
def relu(x):
    return np.maximum(0, x)
```

```python
#plt.subplot(4, 1, 1)
plt.figure(figsize=(5, 3))
plt.plot(x, relu(x), label='relu')
plt.title('relu Activation Function')
plt.xlabel('Input')
plt.ylabel('Output')
plt.grid(True)
plt.tight_layout()
plt.show()
```

# ReLU 계열 - Leaky ReLu

**Leaky ReLU 함수**:
- 수식: 
$$
\text { Leaky } \operatorname{ReLU}(x)= \begin{cases}x & \text { if } x>0 \\ \alpha x & \text { if } x \leq 0\end{cases}
$$

- 설명:   
> ReLU 함수의 변형으로, 입력 값이 음수일 때도 작은 기울기를 유지합니다. 이로 인해 죽은 ReLU 문제를 완화합니다.

```python
def leaky_relu(x, alpha=0.1):
    return np.where(x > 0, x, x * alpha)
```

```python
#plt.subplot(4, 1, 1)
plt.figure(figsize=(5, 3))
plt.plot(x, leaky_relu(x), label='leaky_relu')
plt.title('leaky_relu Activation Function')
plt.xlabel('Input')
plt.ylabel('Output')
plt.grid(True)
plt.tight_layout()
plt.show()
```

# ReLU계열 - PReLU

**Parametric ReLU 함수 (PReLU)**:
- 수식: 
$$
\operatorname{PReLU}(x)= \begin{cases}x & \text { if } x>0 \\ \alpha x & \text { if } x \leq 0\end{cases}
$$

- 설명:   
> Leaky ReLU와 비슷하지만, 음수 입력에 대한 기울기 $\alpha$를 학습합니다.

```python
def prelu(x, alpha=0.05):
    return np.where(x > 0, x, x * alpha)
```

```python
#plt.subplot(4, 1, 1)
plt.figure(figsize=(5, 3))
plt.plot(x, prelu(x), label='prelu')
plt.title('prelu Activation Function')
plt.xlabel('Input')
plt.ylabel('Output')
plt.grid(True)
plt.tight_layout()
plt.show()
```

# ReLU계열 - elu

```python
def elu(x, alpha=1.0):
    return np.where(x > 0, x, alpha * (np.exp(x) - 1))
```

```python
#plt.subplot(4, 1, 1)
plt.figure(figsize=(5, 3))
plt.plot(x, elu(x), label='elu')
plt.title('elu Activation Function')
plt.xlabel('Input')
plt.ylabel('Output')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
def gelu(x):
    return 0.5 * x * (1 + np.tanh(np.sqrt(2 / np.pi) * (x + 0.044715 * x**3)))
```

```python
#plt.subplot(4, 1, 1)
plt.figure(figsize=(5, 3))
plt.plot(x, gelu(x), label='gelu')
plt.title('gelu Activation Function')
plt.xlabel('Input')
plt.ylabel('Output')
plt.grid(True)
plt.tight_layout()
plt.show()
```

# ReLU계열 전체 비교

```python
# ReLU 함수
plt.figure(figsize=(5, 3))
plt.plot(x, relu(x), label='ReLU', color='green')
plt.plot(x, leaky_relu(x), label='Leaky ReLU', color='blue')
plt.plot(x, prelu(x), label='PReLU', color='red')
plt.title('ReLU, Leaky ReLU, and PReLU Activation Functions')
plt.xlabel('Input')
plt.ylabel('Output')
plt.legend()
plt.grid(True)
```

# 시그모이드 함수 (Sigmoid Function)

   - 설명:  
>  입력 값을 0과 1 사이로 변환합니다.  
 (출력층에서)출력이 확률처럼 해석될 수 있기 때문에 이진 분류 문제에 자주 사용됩니다.     
 그러나 큰 입력 값이나 작은 입력 값에 대해 그래디언트 소실 문제가 발생할 수 있습니다.  

```python
# 시그모이드 함수
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
#plt.subplot(4, 1, 1)
plt.figure(figsize=(5, 3))
plt.plot(x, sigmoid(x), label='Sigmoid')
plt.title('Sigmoid Activation Function')
plt.xlabel('Input')
plt.ylabel('Output')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
def sigmoid_derivative(x):
    sx = sigmoid(x)
    return sx * (1 - sx)
```

```python
#plt.subplot(4, 1, 1)
plt.figure(figsize=(5, 3))
plt.plot(x, sigmoid_derivative(x), label='sigmoid_derivative')
plt.title('sigmoid_derivative Activation Function')
plt.xlabel('Input')
plt.ylabel('Output')
plt.grid(True)
plt.tight_layout()
plt.show()
```

# 하이퍼볼릭 탄젠트 함수 (Tanh Function)
- 수식: 
$$
\tanh (x)=\frac{e^x-e^{-x}}{e^x+e^{-x}}
$$
- 출력 범위: (-1, 1)
- 설명:   
> 입력 값을 -1과 1 사이로 변환합니다.   
시그모이드 함수와 비슷하지만 출력이 중심 대칭적이어서 종종 시그모이드 함수보다 더 나은 성능을 보입니다.  
역전파 최대값 차이로 인해 탄젠트가 좀 더 나은 성능을 보임

```python
def tanh(x):
    return np.tanh(x)
```

```python
#plt.subplot(4, 1, 1)
plt.figure(figsize=(5, 3))
plt.plot(x, tanh(x), label='tanh')
plt.title('Sigmoid Activation Function')
plt.xlabel('Input')
plt.ylabel('Output')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
def tanh_derivative(x):
    return 1.0 - np.tanh(x) ** 2
```

```python
#plt.subplot(4, 1, 1)
plt.figure(figsize=(5, 3))
plt.plot(x, tanh_derivative(x), label='tanh_derivative')
plt.title('tanh_derivative Function')
plt.xlabel('Input')
plt.ylabel('Output')
plt.grid(True)
plt.tight_layout()
plt.show()
```

# Softmax

- 수식: 
$$
\operatorname{softmax}\left(x_i\right)=\frac{e^{x_i}}{\sum_j e^{x_j}}
$$

```python
def softmax(x):
    e_x = np.exp(x - np.max(x))
    return e_x / e_x.sum(axis=0)
```

- 가정) 분류문제에서 6개의 클래스로 분류, 3번이 확률값이 제일 높음으로 3번으로 출력
- 소프트맥스의 출력값의 합은 1
- 설명: 여러 클래스 중 하나를 선택하는 다중 클래스 분류 문제에서 사용됩니다. 입력 벡터를 확률 벡터로 변환하여 출력합니다.

```python
# 소프트맥스 함수는 여러 입력 값에 대해 적용되므로 간단한 예시로 시각화
plt.figure(figsize=(10, 6))
softmax_input = np.array([1, 2, 3, 4, 1, 2, 3])
softmax_output = softmax(softmax_input)
plt.bar(range(len(softmax_input)), softmax_output, color='purple')
plt.title('Softmax Activation Function')
plt.xlabel('Class')
plt.ylabel('Probability')
plt.grid(True)

plt.tight_layout()
plt.show()
```
