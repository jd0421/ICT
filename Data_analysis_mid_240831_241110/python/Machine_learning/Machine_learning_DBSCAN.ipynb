# DBSCAN

- 밀도 기반 클러스터링 알고리즘 , 데이터 밀도가 높은 부분을 클러스터로 형성, 밀도가 낮은 부분은 노이즈
- 군집 수를 사전에 지정 할 필요가 없다
- 복잡한 형태의 데이터에서도 클러스터링이 유용

```python
# 라이브러리 임포트
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
```

```python
# 데이터 생성
# ,_ : 두 번째 반환값을 사용하지 않겠다
X,_ = make_moons(n_samples=200, noise=0.05, random_state=0)
```

```python
# 데이터를 정규화 
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

# DBSCAN 클러스터링


```python
dbscan = DBSCAN(eps=0.4 , min_samples=1)
dbscan_labels = dbscan.fit_predict(X_scaled)
```

 - `eps` : 실험적으로 최적의 값을 찾으셔야
 - `min_samples` = 데이터 차원수에 +1 (권장)

```python
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=dbscan_labels, cmap='viridis', s=50)
plt.title('DBSCAN Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

```python
unique_labels = set(dbscan_labels)
print(f"클러스터 수 (노이즈 포함): {len(unique_labels)}")
print(f"노이즈 포인트 수: {list(dbscan_labels).count(-1)}")
```
