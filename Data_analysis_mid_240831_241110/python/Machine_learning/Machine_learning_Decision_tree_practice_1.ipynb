```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
import pandas as pd
```

```python
# 1. 데이터 준비
data = {
    "색깔": ["빨강", "빨강", "초록", "노랑"],
    "크기": ["작음", "큼", "큼", "중간"],
    "과일 이름": ["체리", "사과", "수박", "바나나"]
}

df = pd.DataFrame(data)
```

```python
# 2. 문자열 데이터를 숫자로 변환
color_mapping = {"빨강": 0, "초록": 1, "노랑": 2}
size_mapping = {"작음": 0, "중간": 1, "큼": 2}
fruit_mapping = {"체리": 0, "사과": 1, "수박": 2, "바나나": 3}
```

```python
df["색깔"] = df["색깔"].map(color_mapping)
df["크기"] = df["크기"].map(size_mapping)
df["과일 이름"] = df["과일 이름"].map(fruit_mapping)
```

```python
# 3. 입력 데이터와 레이블 분리
X = df[["색깔", "크기"]]  # 입력 데이터
y = df["과일 이름"]       # 레이블 (타겟)
```

```python
# 4. 의사결정 나무 모델 훈련
model = DecisionTreeClassifier()
model.fit(X, y)
```

```python
# 5. 새로운 데이터로 예측
new_data = [[0, 0]]  # 빨강, 작음
prediction = model.predict(new_data)
```

```python
# 6. 결과 출력
predicted_fruit = {v: k for k, v in fruit_mapping.items()}
print("예측한 과일 이름:", predicted_fruit[prediction[0]])
```

```python
# 7. 나무 구조 시각화 (선택 사항)
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
tree.plot_tree(model, feature_names=["color", "size"], class_names=list(fruit_mapping.keys()), filled=True)
plt.show()
```
