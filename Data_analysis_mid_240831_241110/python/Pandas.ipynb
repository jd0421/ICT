결측치를 어떻게 처리하지??
- 데이터가 충분하다 -> 제거! & 결측치 컬럼이 비율이 적다
- 채워줄수도있다
     - 0 , 평균값, 중앙값, 분산, 빈번값


```python
# 청소년의 결측기 5개를 mean 으로 채워주기

# mean 값 
df['청소년'].mean() #= 463.0388529139685



```

```python
# 결측치 보완
df['청소년'] = df['청소년'].fillna(df['청소년'].mean())
```

```python
df.isnull().sum()
```

```python
df['날씨'].unique()
# ffill(): 결측값을 앞의 값으로 채움
# bfill(): 결측값을 뒤의 값으로 채움
```

```python
df['날씨'] = df['날씨'].bfill()
```
